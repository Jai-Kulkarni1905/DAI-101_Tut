CATEGORICAL ANALYSIS:

```python
sns.countplot(x="transmission", data=carInfo)
plt.title('Transmission types')
plt.show()
```

```python
tags = carInfo['transmission'].value_counts()
tags.plot(kind='pie', autopct='%1.1f%%', label='')
plt.title('Distribution of transmissions')
plt.show()
```

```python
sns.countplot(x='fuelType', data=carInfo)
plt.title('fuell type')
plt.show()
```

```python
carInfo['fuelType'].value_counts()
```

```python
fueltype = carInfo['fuelType']
transmission = carInfo['transmission']
price = carInfo['price']
```

```python
fig = plt.bar(fueltype, price)
plt.show()
```

```python
fig = plt.bar(transmission, price)
plt.show()
```

```python
sns.boxenplot(x=fueltype, y=price)
```

NUMERICAL ANALYSIS:

```python
carInfo['year'].hist(bins=15)
```

```python
sns.displot(carInfo.price)
plt.title("car price distrubution")
plt.show()
```

```python
carInfo[['mileage', 'tax', 'mpg', 'engineSize']].hist(bins=30, figsize=(10,10), color = 'yellowgreen')
```

```python
sns.kdeplot(x = 'year',y = price, cbar = True, data = carInfo, fill= True, color ='red')
plt.show()
```

```python
plt.title('fuel economy vs price')
sns.scatterplot(x = carInfo['mpg'], y = price, hue = transmission)
plt.xlabel("Fuel Economy")
plt.ylabel('Price')
plt.show()
```

```python
carInfo.head()
```

```python
plt.title('fuel economy vs price')
sns.scatterplot(x = carInfo['mpg'], y = price, hue = fueltype)
plt.xlabel("Fuel Economy")
plt.ylabel('Price')
plt.show()
```

```python
sns.countplot(y = 'model', data= carInfo)
plt.title('Models')
plt.show()
```

```python
carInfo.value_counts('model')
```

```python
carInfo.replace({'transmission':{'Manual':0, 'Automatic':1, 'Semi-Auto':2}}, inplace=True)
carInfo.replace({'fuelType':{'Petrol':0, 'Diesel': 1, 'Hybrid': 2, 'Electric':3, 'Other': 4}}, inplace= True)
```

```python
cars=carInfo.drop("model", axis=1)
cars.head()
```

```python
cars.corr()
```

```python
sns.heatmap(cars.corr(), annot = True)
plt.title('Correlation')
plt.show()
```

```python
cars.corr()['price'].sort_values()
```

```python
plt.title('correlation bw year and price')
sns.regplot(x=price, y='year', data = carInfo)
plt.show()
```

```python
pd.plotting.scatter_matrix(cars, diagonal = "kde", figsize=(25, 20))
plt.show()
```

CONSIDERING ALL THE ANALYSIS DONE ABOVE, WE CAN CONCLUDE THAT THE PRICE OF A FORD CAR WOULD DEPEND MOSTLY ON ITS YEAR, FOLLOWED BY ENGINE SIZE AND THEN THE TAXES FOR THAT PARTICULAR CAR!!

THE PRICE DOES NOT DEPEND ON THE MILEAGE OF THE CAR AND THE MPG OF THE CAR.
